```python
# ğŸ”´ Series (ì´ì¤‘ ë¼ë²¨ë§)

## ğŸ”¸ ì¸ë±ìŠ¤ì— ë‹¤ì°¨ì› ë¦¬ìŠ¤íŠ¸(ì•„ì´í…œ 2ê°œ)ë¥¼ ì „ë‹¬
## ğŸ”¸ ë‹¤ì°¨ì›ë¦¬ìŠ¤íŠ¸[0] ìƒìœ„ê³„ì¸µ
## ğŸ”¸ ìƒìœ„ ê³„ì¸µ ì‘ì„±ì‹œ ì£¼ì˜ì  : ê° ê³„ì¸µë³„ë¡œ ì†í•˜ëŠ” í•˜ìœ„ê³„ì¸µ ê°’ì˜ ê°œìˆ˜ë§Œí¼ ê³„ì¸µëª… ì‘ì„±
## ğŸ”¸ ìƒìœ„ ê³„ì¸µ ë¦¬ìŠ¤íŠ¸ì˜ ê°œìˆ˜ = í•˜ìœ„ ê³„ì¸µ ë¦¬ìŠ¤íŠ¸ ê°œìˆ˜
## ğŸ”¸ ë‹¤ì°¨ì›ë¦¬ìŠ¤íŠ¸[1] : í•˜ìœ„ê³„ì¸µ

## ğŸ”¸ ì‹¤ìŠµ ---->
### ë¡œìš° ì¸ë±ìŠ¤ - ìƒìœ„ ê³„ì¸µ : a, b, c, d
### í•˜ìœ„ ê³„ì¸µ - a(1, 2, 3), b(1, 2), c(1, 2, 3, 4), d(1)

s1 = pd.Series(np.arange(10), index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'c', 'c', 'd'],
                                     [1, 2, 3, 1, 2, 1, 2, 3, 4, 1]])

# ğŸ’› ì¸ë±ìŠ¤ë¥¼ ì´ì¤‘ ë¦¬ìŠ¤íŠ¸ë¡œ ì£¼ë©´ ê°™ì€ ê°’ë¼ë¦¬ ë¬¶ì—¬ì„œ ê·¸ë£¹í•‘ì´ ëœë‹¤.
```

```python
# ğŸ”´ í•˜ìœ„ ê³„ì¸µì— ìŠ¬ë¼ì´ì‹± : loc ë©”ì„œë“œë¥¼ ì´ìš©í•´ì„œ ìŠ¬ë¼ì´ì‹±

## ğŸ”¸ ìƒìœ„ ê³„ì¸µì´ cì´ê³  í•˜ìœ„ê³„ì¸µì„ 2 ~ 4 ê¹Œì§€ ìŠ¬ë¼ì´ì‹±(.locëŠ” ë§ˆì§€ë§‰ ë²ˆí˜¸ë„ í¬í•¨)

s1['c'].loc[2:4]
```

```python
# ğŸ”´ unstack() : ìµœí•˜ìœ„(ê¸°ë³¸ë™ì‘)ì— ìˆëŠ” ë¡œìš° ê³„ì¸µì„ ì»¬ëŸ¼ìœ¼ë¡œ ì ìš©í•˜ì—¬ ìœ„ë¡œ ì˜¬ë¦¼

## ğŸ”¸ Series ê°ì²´ë¥¼ DataFrameí™” í•  ìˆ˜ ìˆìŒ
## ğŸ”¸ NaN : ê¸°ì¡´ì— ì—†ë˜ ë¡œìš° ê³„ì¸µì— ëŒ€í•œ ê°’

s1.unstack()

## ğŸ”¸ ì¼ì°¨ì› ë°ì´í„°ë¥¼ 2ì°¨ì› ë°ì´í„°ë¡œ ë°”ê¾¸ì–´ë²„ë¦¼
```

```python
# ğŸ”´ stack() ë©”ì„œë“œ : ì»¬ëŸ¼ì— ìˆë˜ ê°’ì„ ê°€ì¥ í•˜ìœ„ ì¸ë±ìŠ¤ë¡œ ë‚´ë ¤ì„œ ì¬ë°°ì—´

## ğŸ”¸ DataFrameì„ ë‹¤ì‹œ Seriesí™” í•  ë•Œ ì‚¬ìš©

s1.unstack().stack()
```

```python
# ğŸ”´ DataFrame ìƒì„±

## êµ¬ì¡° : 4,5
## row : ìƒìœ„(2017, 2018), í•˜ìœ„(ëª¨ë“  ìƒìœ„ ì¸ë±ìŠ¤ì— ëŒ€í•´ ë™ì¼í•˜ê²Œ a, b)
## col : ìƒìœ„(ì„œìš¸, ê²½ê¸°), í•˜ìœ„(ì„œìš¸- ê°•ë‚¨,ì ì‹¤ / ê²½ê¸°-ì•ˆì–‘,ìˆ˜ì›,íŒêµ)
## ê°’ : 1ì”© ì¦ê°€í•˜ëŠ” 20ê°œ

df = pd.DataFrame(np.arange(20).reshape(4,5), 
                  index=[[2017, 2017, 2018, 2018],['a', 'b', 'a', 'b']], 
                  columns = [['ì„œìš¸', 'ì„œìš¸', 'ê²½ê¸°', 'ê²½ê¸°', 'ê²½ê¸°'], ['ê°•ë‚¨', 'ì ì‹¤', 'ì•ˆì–‘', 'ìˆ˜ì›', 'íŒêµ']])
df
```

```python
# ğŸ”´ ì»¬ëŸ¼ì˜ ìƒìœ„ ê³„ì¸µ ì ‘ê·¼ : ì¼ë°˜ì ì¸ df ì»¬ëŸ¼ ì¸ë±ì‹± ë°©ì‹

## 'ì„œìš¸' ë°ì´í„° ì¡°íšŒ

df['ì„œìš¸']
```

```python
# ğŸ”´ ì»¬ëŸ¼ì´ ì´ì¤‘ êµ¬ì¡°ë¡œ ë˜ì–´ ìˆìœ¼ë¯€ë¡œ, ì¸ë±ì‹±ë„ ì´ì¤‘ìœ¼ë¡œ í•´ì•¼ í•˜ë‚˜ì˜ ìš”ì†Œë¥¼ ì¡°íšŒí•  ìˆ˜ ìˆìŒ

print(df['ì„œìš¸']['ê°•ë‚¨'])
print("--------------------")
print(df['ì„œìš¸', 'ê°•ë‚¨'])
print("--------------------")

## ğŸ”¸ íŠœí”Œë¡œ ì „ë‹¬ì‹œ í•˜ë‚˜ì˜ ë¦¬ìŠ¤íŠ¸ë¡œ 2ê°œ ì´ìƒì˜ ê³„ì¸µ ì²˜ë¦¬ ê°€ëŠ¥
print(df[('ì„œìš¸', 'ê°•ë‚¨')])
```

### â¬› ê°ì²´ ë³µì‚¬

- í• ë‹¹ê¸°í˜¸(=) / ìŠ¬ë¼ì´ì‹± : ì›ë³¸ê³¼ ìƒí˜¸ ì¢…ì†ì ì¸ ë³µì‚¬ë³¸ ê°ì²´ ìƒì„± (ì–•ì€ ë³µì‚¬)
- obj.copy() : ì›ë³¸ê³¼ ë…ë¦½ì ì¸ ë³µì‚¬ë³¸ ê°ì²´ ìƒì„±(ê¹Šì€ ë³µì‚¬)

```python
# ğŸ”´ ëª¨ë“  ê°’ì´ 1ì¸ 5*1 êµ¬ì¡°ì˜ Series ìƒì„±

s1 = pd.Series(np.ones(5))
s1
```

```python
# ğŸ”´ í• ë‹¹ê¸°í˜¸ = ë¡œ ë³µì‚¬ : ì–•ì€ ë³µì‚¬

s2 = s1
s2
```

```python
# ğŸ”´ í• ë‹¹ ê¸°í˜¸ë¡œ ë³µì‚¬í•œ ê°ì²´ì˜ ì•„ì´í…œ ìˆ˜ì •

s2[0] = 10
s2
```

```python
## ğŸ”¸ =ì„ ì´ìš©í•œ ëŒ€ì…ì€ ì›ë³¸ ê°ì²´ë„ ë³€ê²½ë¨
s1
```

```python
# ğŸ”´ ì›ë³¸ ê°ì²´ì˜ ì•„ì´í…œ ìˆ˜ì •

## ğŸ”¸ ì›ë³¸ë„ ë³µì‚¬ë³¸ì—ê²Œ ì˜í–¥ì„ ì¤Œ

s1[1] = 30
print(s1)
print(s2)
```

```python
# ğŸ”´ ìŠ¬ë¼ì´ì‹±ìœ¼ë¡œ ë³µì‚¬

## ğŸ”¸ ì¼ë°˜ íŒŒì´ì¬ì—ì„œëŠ” ê¹Šì€ ë³µì‚¬ë¡œ ê°„ì£¼ë˜ì§€ë§Œ
## ğŸ”¸ íŒ¬ë”ìŠ¤ì—ì„œëŠ” ì–•ì€ ë³µì‚¬ë¡œ ì·¨ê¸‰ëœë‹¤. ğŸ’›

s3 = s1[:]
s3
```

```python
## ğŸ”¸ ë³µì‚¬ë³¸ ê°ì²´ ì•„ì´í…œ ìˆ˜ì • => s1, s2ì— ì˜í–¥

s3[2] = -9

print("------< s3 >----")
print(s3)
print("------< s1 >----")
print(s1)
print("------< s2 >----")
print(s2)
```

```python
# ğŸ”´ ì›ë³¸ ë³€ê²½ => s2, s3ì— ì˜í–¥

s1[3] = 33

print("------< s1 >----")
print(s1)
print("------< s2 >----")
print(s2)
print("------< s3 >----")
print(s3)
```

```python
# ğŸ”´ ê¹Šì€ ë³µì‚¬ë¥¼ í•˜ê³ ì‹¶ìœ¼ë©´ ? => .copy()ë¥¼ ì´ìš©

s4 = s1.copy()
s4
```

```python
## ğŸ”¸ s4ë¥¼ ë°”ê¿”ë„ s1, s2, s3ì€ ì˜í–¥ì„ ë°›ì§€ ì•ŠìŒ.

s4[4] = 200

print("------< s1 >----")
print(s1)
print("------< s2 >----")
print(s2)
print("------< s3 >----")
print(s3)
print("------< s4 >----")
print(s4)
```

```python
## ğŸ”¸ ë§ˆì°¬ê°€ì§€ë¡œ s1ì„ ë³€ê²½í•´ë„ s4ì—ëŠ” ì˜í–¥ì„ ì£¼ì§€ ì•ŠìŒ

s1[0] = 2929229

print("------< s1 >----")
print(s1)
print("------< s2 >----")
print(s2)
print("------< s3 >----")
print(s3)
print("------< s4 >----")
print(s4)
```

```python
# ğŸ”´ ì‹œë¦¬ì¦ˆ ìƒì„±

## ê°’ê³¼ ì¸ë±ìŠ¤ë¼ë²¨ì´ ìˆœì„œëŒ€ë¡œ ë“¤ì–´ê°€ì§€ ì•Šì€ ì‹œë¦¬ì¦ˆ

s1 = pd.Series([2, 3, 1, 7, 0], index=list('gacfd'))
s1
```

```python
## ğŸ”¸ ì¼ì‹œì  ë³€ê²½ì´ë¯€ë¡œ s1ì—ëŠ” ì˜í–¥ì„ ì£¼ì§€ ì•ŠìŒ

s1
```

```python
# ğŸ”´ ì»¬ëŸ¼ëª… ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬

df1.sort_index(axis=1)
```

```python
# ğŸ”´ ì»¬ëŸ¼ëª… ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬

df1.sort_index(axis=1, ascending=False)
```

```python
# ğŸ”´ ì»¬ëŸ¼ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬ í›„, ë¡œìš° ê¸°ì¤€ìœ¼ë¡œ ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬

df1.sort_index(axis=1, ascending=False).sort_index()
```
