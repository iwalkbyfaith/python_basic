```python
# üî¥ 1. ÏúÑÏùò urlÏùÑ Ïù¥Ïö©Ìï¥ Îç∞Ïù¥ÌÑ∞Î•º ÏöîÏ≤≠Ìï©ÎãàÎã§. (getJSONÍ≥º ÎπÑÏä∑Ìïú ÎäêÎÇåÏù¥ÎùºÍ≥† ÏÉùÍ∞ÅÌïòÎ©¥ Îê®)

r = urllib.request.urlopen(url)
```

```python
# üî¥ 2. ÏöîÏ≤≠ ÌéòÏù¥ÏßÄÏùò Í≤∞Í≥º Îç∞Ïù¥ÌÑ∞Î•º ÌååÏù¥Ïç¨ ÎÇ¥Î∂Ä Îç∞Ïù¥ÌÑ∞Î°ú Î∞îÍøâÎãàÎã§.

json_raw = r.read()
```

```python
# üî¥ 3. Ï†úÏù¥Ïä® Îç∞Ïù¥ÌÑ∞Í∞Ä Î∞îÏù¥ÎÑàÎ¶¨ ÌÉÄÏûÖÏúºÎ°ú Îì§Ïñ¥Ïò§Îäî Í≤ÉÏùÑ Î≥º Ïàò ÏûàÏäµÎãàÎã§.
json_raw
```

```python
## üî∏ ÌòÑÏû¨Îäî Î∞îÏù¥Ìä∏ ÏûêÎ£åÌòïÏûÑ => Î¨∏ÏûêÎèÑ ÏïÑÎãàÍ≥† jsonÎèÑ ÏïÑÎãò
## üî∏ print Ìï¥Î≥¥Î©¥ b'~~~~~'Î°ú ÏãúÏûëÌïòÎäî Îç∞Ïù¥ÌÑ∞. 
## üî∏ (Î¨∏Ï†ú) Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏúºÎ°ú Î∞îÎ°ú Î≥ÄÌôòÏù¥ Î∂àÍ∞ÄÎä•ÌïòÎã§.

type(json_raw)
```

```python
# üî¥ (Ìï¥Í≤∞) 4. utf-8 ÌòïÏãùÏúºÎ°ú Í≥†Ï≥êÏ§òÏïºÌïúÎã§.

## üî∏Î¨∏Î≤ïüî∏ .decode()Î•º Ïù¥Ïö©
## üî∏ encode => Ïö∞Î¶¨Í∞Ä Ïì∞Îäî Î¨∏Ïûê->Ïª¥Ìì®ÌÑ∞Í∞Ä Ïì∞Îäî ÌòïÌÉú
## üî∏ decode => Ïª¥Ìì®ÌÑ∞Í∞Ä Ïì∞Îäî ÌòïÌÉúÏùò Î¨∏Ïûê -> Ïö∞Î¶¨Í∞Ä Ïì∞Îäî ÌòïÌÉúÎ°ú Î≥ÄÌôò

json_utf8 = json_raw.decode("utf-8")
json_utf8
```

```python
## üî∏ Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ ÌôïÏù∏ (ÏûêÎ£åÌòï : ÎîïÏÖîÎÑàÎ¶¨ Íµ¨Ï°∞Î•º Ïú†ÏßÄÌïòÍ≥† ÏûàÎäî Î¨∏Ïûê)

type(json_utf8)
```

```python
# üî¥ 5. ÎîïÏÖîÎÑàÎ¶¨ ÌòïÌÉúÎ°ú Ï†ÅÌûå 'Î¨∏ÏûêÏó¥ ÏûêÎ£å'Î•º 'ÎîïÏÖîÎÑàÎ¶¨' ÌòïÌÉúÎ°ú Î≥ÄÌôò
## üî∏Î¨∏Î≤ïüî∏ json.loads(ÎîïÏÖîÎÑàÎ¶¨ ÌòïÌÉúÏùò Î¨∏ÏûêÏó¥ ÏûêÎ£å)

json_complete = json.loads(json_utf8)
```

```python
# üî¥ 6. ÏµúÏ¢Ö Î≥ÄÌôò ÏôÑÎ£å
json_complete
```

```python
## üî∏ ÌÉÄÏûÖ ÌôïÏù∏
type(json_complete)
```

```python
# üî¥ 7. ÎîïÏÖîÎÑàÎ¶¨ ÏûêÎ£åÎ•º Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÌòïÌÉúÎ°ú Î≥ÄÌôòÌï¥Ï§å.

## üî∏Î¨∏Î≤ïüî∏ json_normalize(ÎîïÏÖîÎÑàÎ¶¨ ÏûêÎ£å)

pd.json_normalize(json_complete)
```

```python
# üî¥ 8. item Ïª¨ÎüºÏùÑ ÌÉÄÍ≤üÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÌôî Ìï©ÎãàÎã§

json_complete['item']
```

```python
# üî¥ 9. Ïπ¥ÌÖåÍ≥†Î¶¨ Î≥Ñ Î≤†Ïä§Ìä∏ ÏÖÄÎü¨ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÌôîÌïòÍ∏∞

box_result = pd.json_normalize(json_complete['item'])
box_result
```

```python
# üî¥ 10. Ïπ¥ÌÖåÍ≥†Î¶¨ Î≥ÑÎ°ú Í∞ÄÏ†∏Ïò§Í∏∞ 100, 200, 300ÍπåÏßÄÎßå Ìï¥Î≥¥Í∏∞

## üî∏ 1. Ïπ¥ÌÖåÍ≥†Î¶¨ ÏÑ§Ï†ï
target_category = ['101', '102', '130']

## üî∏ 2. Îπà Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
best_seller_result = pd.DataFrame()

for category in target_category :
    print(category + "ÏßÑÌñâÏ§ë")
    url = "http://book.interpark.com/api/bestSeller.api?key=%s&categoryId=%s&output=json" %(api_key, category)
    
    # ÏúÑÏùò url Î≥ÄÏàòÎ•º Ïù¥Ïö©Ìï¥ Îç∞Ïù¥ÌÑ∞Î•º ÏöîÏ≤≠
    r = urllib.request.urlopen(url)
    
    # ÏöîÏ≤≠ ÌéòÏù¥ÏßÄÏùò Í≤∞Í≥º Îç∞Ïù¥ÌÑ∞Î•º ÌååÏù¥Ïç¨ ÎÇ¥Î∂Ä Îç∞Ïù¥ÌÑ∞Î°ú Î∞îÍøà
    json_raw = r.read()
    json_utf8 = json_raw.decode("utf-8")
    json_complete = json.loads(json_utf8)
    pd.json_normalize(json_complete)
    result = pd.json_normalize(json_complete['item'])
    
    # Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùÑ ÌïòÎÇòÎ°ú ÌÜµÌï©
    best_seller_result = best_seller_result.append(result)
    
    # Ïó¨Îü¨Í∞ú Í∞ÄÏ†∏Ïò¨ ÎïåÎäî time.sleep Í±∏Î©¥ÏÑú ÌïòÍ∏∞
    
    
best_seller_result.to_excel('C:/python_cr/best_seller_result_test5.xlsx')
```

# ‚¨õ best_seller_result Î≥ÄÏàòÏóê Îç∞Ïù¥ÌÑ∞Î•º ÎÑ£Í≥†, ÏóëÏÖÄ ÌååÏùºÎèÑ ÎßåÎì§Í∏∞

```python
## üî∏ 1. Ïπ¥ÌÖåÍ≥†Î¶¨ ÏÑ§Ï†ï

target_category = []

target_category.extend(list(range(101, 130)))
target_category.extend(list(range(201, 218)))
target_category.extend(list(range(301, 321)))
target_category.extend(list(range(409, 431)))

## üî∏ 2. Îπà Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
best_seller_result = pd.DataFrame()

for category in target_category :
    print(str(category) + "ÏßÑÌñâÏ§ë")
    url = "http://book.interpark.com/api/bestSeller.api?key=%s&categoryId=%s&output=json" %(api_key, category)
    
    # ÏúÑÏùò url Î≥ÄÏàòÎ•º Ïù¥Ïö©Ìï¥ Îç∞Ïù¥ÌÑ∞Î•º ÏöîÏ≤≠
    r = urllib.request.urlopen(url)
    
    # ÏöîÏ≤≠ ÌéòÏù¥ÏßÄÏùò Í≤∞Í≥º Îç∞Ïù¥ÌÑ∞Î•º ÌååÏù¥Ïç¨ ÎÇ¥Î∂Ä Îç∞Ïù¥ÌÑ∞Î°ú Î∞îÍøà
    json_raw = r.read()
    json_utf8 = json_raw.decode("utf-8")
    json_complete = json.loads(json_utf8)
    pd.json_normalize(json_complete)
    result = pd.json_normalize(json_complete['item'])
    
    # Ïπ¥ÌÖåÍ≥†Î¶¨Î≥Ñ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑÏùÑ ÌïòÎÇòÎ°ú ÌÜµÌï©
    best_seller_result = best_seller_result.append(result)
    
    # Ïó¨Îü¨Í∞ú Í∞ÄÏ†∏Ïò¨ ÎïåÎäî time.sleep Í±∏Î©¥ÏÑú ÌïòÍ∏∞
    time.sleep(3)
    
best_seller_result.to_excel('C:/python_cr/best_seller_result.xlsx')
```
