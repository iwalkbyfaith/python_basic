```python
# ğŸ”´ (ì‹¤ìŠµ) ë‚ ì§œë³„ë¡œ ê°€ì ¸ì˜¤ê¸°

## ğŸ”¸ ['20220620', '20220621', '20220622']
## ğŸ”¸ ë°˜ë³µë¬¸ê³¼ appendë¥¼ ì´ìš©í•´ì„œ 3ì¼ì¹˜ ë°ì´í„°ë¥¼ ìš”ì²­í•´ì„œ í•©ì³ë³´ì„¸ìš”
## ğŸ”¸ ë¹ˆ ë°ì´í„° í”„ë ˆì„ì„ í•˜ë‚˜ ìƒì„±í•´ë†“ê¸°



target_dates = ['20220620', '20220621', '20220622'] # <<- ë‚ ì§œë¥¼ í•˜ë‚˜í•˜ë‚˜ ì…ë ¥í–ˆë‹¤ë©´ ì´ì œëŠ” ë‹¤ë¥¸ ë°©ë²•ì„ ì‚¬ìš©
## ë¨¼ì € ìœ„ì˜ ì…€ì— datetimeì„ import í•œë‹¤.

box_result2 = pd.DataFrame()

for date in target_dates :
    print(date)
    url = "http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=%s&targetDt=%s" % (api_key, date)
    
    # ìœ„ì˜ url ë³€ìˆ˜ë¥¼ ì´ìš©í•´ ë°ì´í„°ë¥¼ ìš”ì²­
    r = urllib.request.urlopen(url)
    
    # ìš”ì²­ í˜ì´ì§€ì˜ ê²°ê³¼ ë°ì´í„°ë¥¼ íŒŒì´ì¬ ë‚´ë¶€ ë°ì´í„°ë¡œ ë°”ê¿ˆ
    json_raw = r.read()
    json_utf8 = json_raw.decode("utf-8")
    json_complete = json.loads(json_utf8)
    pd.json_normalize(json_complete)
    result = pd.json_normalize(json_complete['boxOfficeResult']['dailyBoxOfficeList'])
    
    # ì–»ì–´ì˜¨ ë°ì´í„°ì— ë‚ ì§œ ì»¬ëŸ¼ ì¶”ê°€
    result['target_date'] = date
    print(result)
    
    # ë‚ ì§œë³„ ë°ì´í„°í”„ë ˆì„ì„ í•˜ë‚˜ë¡œ í†µí•©
    box_result2 = box_result2.append(result)
    # í˜¹ì€ concat ì‚¬ìš© : box_result = pd.concat([box_result, result])
    
    # 2ì´ˆê°„ ìœ ì§€
    time.sleep(2)
    
    
    
```
