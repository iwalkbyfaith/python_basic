# ⬛ Json 데이터를 팬더스 데이터 프레임으로 변환

```python
# 🔴 1. 위의 url을 이용해 데이터를 요청합니다. (getJSON과 비슷한 느낌이라고 생각하면 됨)

r = urllib.request.urlopen(url)
```

```python
# 🔴 2. 요청 페이지의 결과 데이터를 파이썬 내부 데이터로 바꿉니다.

json_raw = r.read()
```

```python
# 🔴 3. 제이슨 데이터가 바이너리 타입으로 들어오는 것을 볼 수 있습니다.
json_raw
```

```python
## 🔸 현재는 바이트 자료형임 => 문자도 아니고 json도 아님
## 🔸 print 해보면 b'~~~~~'로 시작하는 데이터. 
## 🔸 (문제) 데이터 프레임으로 바로 변환이 불가능하다.

type(json_raw)
```

```python
# 🔴 (해결) 4. utf-8 형식으로 고쳐줘야한다.

## 🔸문법🔸 .decode()를 이용
## 🔸 encode => 우리가 쓰는 문자->컴퓨터가 쓰는 형태
## 🔸 decode => 컴퓨터가 쓰는 형태의 문자 -> 우리가 쓰는 형태로 변환

json_utf8 = json_raw.decode("utf-8")
json_utf8
```

```python
## 🔸 데이터 타입 확인 (자료형 : 딕셔너리 구조를 유지하고 있는 문자)

type(json_utf8)
```

```python
# 🔴 5. 딕셔너리 형태로 적힌 '문자열 자료'를 '딕셔너리' 형태로 변환
## 🔸문법🔸 json.loads(딕셔너리 형태의 문자열 자료)

json_complete = json.loads(json_utf8)
```

```python
# 🔴 6. 최종 변환 완료
json_complete
```

```python
## 🔸 타입 확인
type(json_complete)
```

```python
# 🔴 7. 딕셔너리 자료를 데이터프레임 형태로 변환해줌.

## 🔸문법🔸 json_normalize(딕셔너리 자료)

pd.json_normalize(json_complete)
```

```python
# 🔴 8. item 컬럼을 타겟으로 데이터 프레임화 합니다

json_complete['item']
```

```python
# 🔴 9. 카테고리 별 베스트 셀러 데이터 프레임화하기

box_result = pd.json_normalize(json_complete['item'])
box_result
```

# ⬛ best_seller_result 변수에 데이터를 넣고, 엑셀 파일도 만들기

```python
## 🔸 1. 카테고리 설정

target_category = []

target_category.extend(list(range(101, 130)))
target_category.extend(list(range(201, 218)))
target_category.extend(list(range(301, 321)))
target_category.extend(list(range(409, 431)))

## 🔸 2. 빈 데이터 프레임 생성
best_seller_result = pd.DataFrame()

for category in target_category :
    print(str(category) + "진행중")
    url = "http://book.interpark.com/api/bestSeller.api?key=%s&categoryId=%s&output=json" %(api_key, category)
    
    # 위의 url 변수를 이용해 데이터를 요청
    r = urllib.request.urlopen(url)
    
    # 요청 페이지의 결과 데이터를 파이썬 내부 데이터로 바꿈
    json_raw = r.read()
    json_utf8 = json_raw.decode("utf-8")
    json_complete = json.loads(json_utf8)
    pd.json_normalize(json_complete)
    result = pd.json_normalize(json_complete['item'])
    
    # 카테고리별 데이터 프레임을 하나로 통합
    best_seller_result = best_seller_result.append(result)
    
    # 여러개 가져올 때는 time.sleep 걸면서 하기
    time.sleep(3)
    
best_seller_result.to_excel('C:/python_cr/best_seller_result.xlsx')
```
