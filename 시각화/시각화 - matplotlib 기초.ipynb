### 스타일 지정

- plt.plot([x], y, [fmt], **kwargs)
    - fmt : '[color][marker][line]'
        - 색상, 마커, 선 등에 대한 포맷 옵션 설정

1) 마커 참조 도큐먼트 주소<br>
https://matplotlib.org/3.1.3/api/markers_api.html<br>
2) 색상 참조 도큐먼트 주소<br>
https://matplotlib.org/3.1.0/gallery/color/named_colors.html<br>
3) 라인 스타일 참조 도큐먼트 주소<br>
https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html<br>

```python
# 스타일 문자열 추가 : [fmt] => color + marker + line 순서대로 문자열 작성
# 생략된 요소는 기본값으로 적용
# 'o' : 라인없이 큰 점(원모양)으로 데이터 표시
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], 'o')
plt.show()
```

```python
# 마커는 동그라미, 라인은 실설
# fmt[색상][마커][선종류]
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], 'o-')
plt.show()
```

```python
# 위 값에 마커는 동그라미, 선은 점선으로 줘 보세요.
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], 'o--')
plt.show()
```

```python
# ro: : r(빨간색), o(마커), :(dotted)
plt.plot([1, 2, 3, 4], [2, 6, 10, 7], 'ro:')
plt.show()
```

```python
# 기타 스타일 파라미터 종류
# color : 선 색상(magenta)
# linestyle : 선의 생김새(-.)
# marker : 각 포인트마다 표시할 도형(^)
# markerfacecolor : 마커에 적용할 색상(orange)
# markersize : 마커 크기(20)
# markeredgecolor : 마커테두리색(r)
plt.plot([1, 2, 3, 4], [2, 6, 10, 7],
        color='magenta', linestyle='-.', marker='^',
        markeredgecolor='r', markerfacecolor='orange', markersize='20')
plt.show()
```

### 축 범위 지정

```python
# x축 최소값 = 0, x축 최대값 = 7
# y축 최소값 = 2, y축 최대값 = 20
plt.plot([1, 2, 3, 4], [10, 3, 8, 13])
plt.axis([0, 7, 2, 20])
plt.show()
```

```python
# 선에 비해 지나치게 작은 플롯을 구성하면 일부 플롯이 잘려나간다
plt.plot([1, 2, 3, 4], [10, 3, 7, 13])
plt.axis([2, 4, 4, 10])
plt.show()
```

```python
# x축, y축 범위 설정(최소값, 최대값)
# x data : 10, 20, 30, 40
# y data : 1, 4, 9, 16
# 색상 : c = magenta
# 선굵기(lw: line width) : 5
# 선모양(ls : line style) : :
# 마커 : o
# 마커 크기(ms : marker size) : 10
# x축 범위 : 0, 50
# y축 범위 : -10, 30

plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c='magenta',
        lw = 5, ls=':', marker='o', ms=10)
# plt.axis([0, 50, -10, 30])
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

```python
# grid 표시하기
# grid는 격자무늬로 표시할때는 plt.grid(True)를 추가로 한 줄 더 입력해주시면 됩니다.
plt.plot([10, 20, 30 ,40], [1, 4, 9, 16], c='magenta', lw=5, ls=":", marker='o', ms=10)
plt.xlim(0, 50)
plt.ylim(-10, 30)
# 아래 차트 제목을 "그리드 적용한 플롯"으로 붙여주세요
plt.title('그리드 적용한 플롯')

plt.grid(True)
plt.show()
```

#### matplotlib.pyplot.scatter 

- 대표값을 중심으로 데이터가 흩어져 있는 정도를 시각화
- 데이터의 분산 정도, 데이터 간의 상관관계를 파악할 수 있음
- plt.scatter(x, y, s=None, c=None, data=None)
    - s : 마커 사이즈
    - c : 마커 색상
    - data : dictionary 자료형을 사용하여 데이터를 전달하면 keyword string을 통해 데이터 사용 가능

```python
# 정규분포 확률밀도에서 100개의 표본 추출
# random.normal()
# 난수값 고정
np.random.seed(123)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.scatter(x, y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
# 딕셔너리 타입으로 데이터 전달
# 샘플 의미 : 데이터마다 색상, 마커크기가 다르게 표현되기 위하여 생성
# x데이터(a) : 0이상 50미만의 범위에서 1씩 증가하는 정수
# 마커색상 데이터(b) : 0이상 50미만의 정수중 50개의 난수 추출
# 마커크기 데이터(s) : 정규분포에서 추출한 50개의 난수(c) -> 발생한 난수의 절대값
# 을 취한 다음 거기에 100배
# y데이터(d) : x데이터 + 10 * 정규분포에서 추출한 50개의 난수
data = {'a':np.arange(0, 50),
       'b':np.random.randint(0, 50, 50),
       'c':np.random.randn(50)}

data['d'] = data['a'] + 10 * data['c']
data['s'] = np.abs(data['c']) * 100
data
```

```python
# 데이터 확인 : x에 해당하는 데이터
data['a']
```

```python
# 데이터 확인 : y에 해당하는 데이터
data['d']
```

```python
# x, y를 각 'a', 'd'로 설정해서 산점도를 그려보세요
plt.scatter(data['a'], data['d'])
plt.show()
```

```python
# c 파라미터에는 수치값을 넣으면 숫자에 매칭된 색깔이 마커에 부여된다.
plt.scatter(data['a'], data['d'], c = data['b'])
plt.show()
```

```python
# s파라미터를 이용해 마커의 크기도 조절 가능합니다. s='s' 형식으로 넣어주세요.
plt.scatter(data['a'], data['d'], c = data['b'], s = data['s'])
plt.show()
```

### Bar chart (막대 그래프)

- 여러 값의 상대적인 크기를 비교하기에 적합한 표현법이며 원형 차트보다 더욱 일반적으로 사용됨
- plt.bar(x, y)
    - x 데이터가 카테고리 값인 경우에는 bar, barh 로 바 차트 시각화 표현
    - 가로 방향으로 바 차트를 그리려면 barh
    = barh 차트 주의점 : x축이 일반적으로 y축의 위치를 의미

```python
# 범주형 데이터에 대해 bar graph 기분
x = ['가', '나', '다']
y = [2, 10, 5]
plt.bar(x, y)
plt.show()
```

```python
# 제목, xlabel, ylabel
plt.bar(x, y)
plt.title('바 차트 테스트')
plt.xlabel('x축데이터')
plt.ylabel('y축데이터')
# 만약 x축 이름을 교체하고 싶다면, xticks를 사용해주세요
plt.xticks(x, ['김', '이', '박'])
plt.show()
```

```python
# 각각의 bar마다 색상 변경
# plt.bar()를 리스트로 받아서 
# 인덱싱 후 .set_color('색상')을 써주시면 됩니다.
bar_list = plt.bar(x, y)
bar_list[0].set_color('r')
bar_list[2].set_color('g')
plt.title('바 차트')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

#### 가로형 막대그래프

- plt.barh(y, x)

```python
# 0번째는 magenta색으로, 1번째는 orange로 해주세요.
bar_list = plt.barh(x, y)
bar_list[0].set_color('magenta')
bar_list[1].set_color('orange')
plt.title('가로형 그래프')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

### pie char

- plt.pie(ratio, labels=, color=, autopct=, shadow=T/F)

### histogram(히스토그램)

- 데이터 분포의 특징, 이상치 여부를 확인하는데 용이
    - 데이터의 분포 
        - 데이터 값의 전반적으로 퍼져있는 형태
        - 데이터의 수가 많을 때 데이터를 파악하기 위해 어떤 값 근처에 어떤 모양으로 모여 있는지 이해하기 쉬움
- 데이터에 이상치가 존재하는 경우 비정상적인 히스토그램이 생성되며 이상치를 제거해주는 작업이 필요
    - 예) 전반적인 데이터 분포가 0 ~ 1 사이에 있는데 한 개의 데이터가 100인 경우 정상적인 히스토그램이 그려지지 않음.

- 히스토그램
    - 자료 값이 가질 수 있는 범위를 몇 개의 구간으로 나누고 각 구간에 해당하는 값의 숫자 혹은 상대적 빈도를 계산하여 시각화
    - plt.hist(data, bins)
        - 일반적인 경우 히스토그램의 구간을 나눌 때 동일한 간격으로 나누기 때문에 구간(bins)의 개수가 중요
        - bins
            - 히스토그램 구간 수를 나타내는 값
            - 구간 수가 너무 작으면(=간격이 너무 넓으면) 세부 분포 파악이 어렵고, 구간 수가 너무 많으면 전체적인 분포 파악이 어려움

```python
# 정규분포에서 무작위로 1000개 데이터 추출
# 히스토그램은 구간이 많고 모집단이 클수록 정규분포에 가깝게 표현된다
np.random.seed(200)
x = np.random.randn(1000)
plt.hist(x, bins=10) # None vs 300 vs 100 vs 50
plt.show()
```

### Box-plot
- box-and-whisker plot
- 최댓값, 최솟값, 사분위수(Q1, Q2, Q3)를 사용하여 자료의 측정값들의 분포와 극단값에 대해 파악
    - IQR(Inter Quartile Range) : Q3 - Q1
    - 최대값 : Q1 + 1.5 x IQR 범위 안에 있는 최대값
    - 최소값 : Q3 - 1.5 x IQR 범위 안에 있는 최대값
    - 박스(Box) : Q1 ~ Q3의 값
    - n개의 데이터가 있을 때 사분위수
        - 중앙값(Q2) : (n + 1) / 2
        - 사분위수
            - 집한 자료를 크기순으로 나열하여 사등분하는 척도
            - (Q2 + 1) / 2를 기준으로 작은 쪽에서 해당 위치에 있는 수를 아래사분위수(Q1), 큰 쪽에 위치한 수를 위사분위수(Q2=n+1-값)
            - 데이터 개수가 홀수일 때는 앞뒤 값의 평균을 사용
- 두 개 이상의 자료 집단을 비교할 때 유용
- 해당 범위에 벗어나는 값은 이상치로 판정하여 표기
