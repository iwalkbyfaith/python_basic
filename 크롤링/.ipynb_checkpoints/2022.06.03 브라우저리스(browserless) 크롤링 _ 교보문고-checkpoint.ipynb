### ğŸŸ¦ (ì‹¤ìŠµ) êµë³´ë¬¸ê³  ìˆœíšŒ í¬ë¡¤ë§ í•˜ê¸°

- êµë³´ë¬¸ê³  ìˆœíšŒí¬ë¡¤ëŸ¬ë¥¼ ì´ìš©í•´ 1 ~ 200ìœ„ê¹Œì§€ì˜ ë„ì„œ ì œëª©, ê°€ê²©, ì €ìë¥¼ ì–»ì–´ì™€ì„œ txt & csv íŒŒì¼ë¡œ ë§Œë“¤ì–´ì£¼ì„¸ìš”

```python
# 0. ë¦¬ìŠ¤íŠ¸ ìƒì„±

total_title = []
total_author = []
total_price = []
```

```python
# 1. êµë³´ë¬¸ê³  ë² ìŠ¤íŠ¸ í˜ì´ì§€ ì ‘ì†

for i in range(1, 11) :
    
    # í•´ë‹¹ í˜ì´ì§€ ì ‘ì† í›„, ì†ŒìŠ¤ì½”ë“œ íŒŒì´ì¬ ë³€ìˆ˜ë¡œ ì €ì¥í•˜ê¸°
    res = requests.get("https://www.kyobobook.co.kr/bestSellerNew/bestseller.laf?targetPage=%s" % i)
    time.sleep(2)
    source = res.text
    
    # ë””ë²„ê¹…ìš© ì½”ë“œ
    print("%sí˜ì´ì§€ ì§„í–‰ì¤‘" %i)
    
    # ì €ì¥ëœ íŒŒì´ì¬ ì½”ë“œ íŒŒì‹±í•˜ê¸°
    parsed_source = BeautifulSoup(source, "html.parser")
    
    detail_list = parsed_source.find_all("div", class_="detail")
    
    # ì œëª©, ì €ì, ê°€ê²©
    for item in detail_list :
        title = item.find_all("div", class_="title")
        total_title.append(title[0].text.strip().replace(",",""))
        # ğŸŸ¡ ì½¤ë§ˆë¥¼ ë¹¼ì§€ ì•Šê³  ì‹¶ë‹¤ë©´, titleì˜ ì• ë’¤ì— ë”°ì˜´í‘œë¥¼ ë¶™ì—¬ì£¼ë©´ ëœë‹¤ ex '"' + title[0].text.strip() + '"'
    
        author = item.find_all("div", class_="author")
        #print(author[0].text.strip().replace("\t","").replace("\r\n","").replace("\n",""))
        total_author.append(author[0].text.strip().replace("\t","").replace("\r\n","").replace("\n","")) 
    
        price = item.find_all("strong", class_="book_price")
        #print(price[0].text.replace(",","").replace("ì›", "") + "\n")
        total_price.append(price[0].text.replace(",","").replace("ì›", ""))
        
    time.sleep(2)
    
```
